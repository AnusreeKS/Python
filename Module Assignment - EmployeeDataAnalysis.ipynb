# Employee Data Analysis - ABC Company

```python
# Importing necessary libraries
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data set
```

```python
df = pd.read_csv("ABC Employee Details.csv")
```

```python
df
```

### Preprocessing: Replace height values with random numbers between 150 and 180

```python
df["Height"] = np.random.randint(150, 181, size=len(df))
```

```python
df
```

## Data Analysis

### 1.Employee Distribution by Team

```python
team_distribution= df["Team"].value_counts()
team_distribution
```

```python
len(df)
```

```python
team_percentage_slpit = ( team_distribution / len(df) )*100
team_percentage_slpit
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x= team_distribution.index, y= team_distribution.values, palette = "viridis")
plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x= team_distribution.index, y= team_percentage_slpit.values, palette = "viridis")
plt.title("Percentage of Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Percentage Split")
plt.xticks(rotation=90)
plt.show()
```

### 2. Segregation by position

```python
position_distribution = df["Position"].value_counts()
plt.figure(figsize=(8,5))
sns.barplot(x=position_distribution.index, y=position_distribution.values, palette="coolwarm")
plt.title("Employee Distribution by Position")
plt.xlabel("Position")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### 3. Predominant Age Group

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=10, kde=True, color="blue")
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
age_distri=df["Age"].value_counts()
```

```python
age_distri
```

### 4. Highest Salary Expenditure (Team and Position)

```python
salary_by_team = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
salary_by_position = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)
```

```python
highest_salary_team = df.groupby("Team")["Salary"].sum().idxmax()
print(f"Team with highest salary expenditure: {highest_salary_team}")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=salary_by_team.index, y=salary_by_team.values, palette="magma")
plt.title("Total Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Total Salary")
plt.xticks(rotation=90)
plt.show()
```

```python
highest_salary_position = df.groupby("Position")["Salary"].sum().idxmax()
print(f"Position with highest salary expenditure: {highest_salary_position}")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=salary_by_position.index, y=salary_by_position.values, palette="cividis")
plt.title("Total Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary")
plt.xticks(rotation=45)
plt.show()
```

### 5. Correlation between Age and Salary

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.6)
plt.title("Correlation between Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
